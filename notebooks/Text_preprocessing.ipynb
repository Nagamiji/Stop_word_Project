# II. Data Preprocessing
 1. Splited the text to sentences

```python
def split_sentences(text):
    for p in "។៕!?":
        text = text.replace(p, p + "<STOP>")
    sentences = text.split("<STOP>")
    sentences = [s.strip() for s in sentences if s.strip()]
    return sentences

# Store as list
all_sentences = []
with open(dataset_text, "r", encoding="utf-8") as f:  # Read file line by line (memory-efficient)
    for line in f:
        line = line.strip()
        if line:
            sentences = split_sentences(line)
            all_sentences.extend(sentences)
# Preview first 10 sentences
print("First 100 sentences:")
for i, sent in enumerate(all_sentences[:100], 1):
    print(f"{i}: {sent}")
```

2. Kept only khmer text which is in dataset

```python
def keep_only_khmer(text):
    allowed_punct = "។៕!?៖ "  # Khmer punctuation + space
    cleaned = "".join([c for c in text if '\u1780' <= c <= '\u17FF' or c in allowed_punct])
    cleaned = " ".join(cleaned.split())  # remove extra spaces
    return cleaned

# Apply cleaning to all sentences
sentences_khmer = [keep_only_khmer(s) for s in all_sentences if keep_only_khmer(s)]

print("First 10 Khmer-only sentences:")
for i, sent in enumerate(sentences_khmer[:100], 1):
    print(f"{i}: {sent}")
```

3. Removed all number from dataset

```python
def remove_khmer_numbers(text):
    khmer_numbers = "០១២៣៤៥៦៧៨៩"
    cleaned = "".join([c for c in text if c not in khmer_numbers])
    return cleaned

sentences_cleaned = []
for s in all_sentences:
    s = keep_only_khmer(s)         # keep only Khmer letters + punctuation
    s = remove_khmer_numbers(s)    # remove Khmer numbers
    s = s.strip()                   # remove extra spaces
    if s:                           # skip empty sentences
        sentences_cleaned.append(s)

print("First 100 fully cleaned sentences:")
for i, sent in enumerate(sentences_cleaned[:100], 1):
    print(f"{i}: {sent}")
```

4. Remove some Punctuations from text

```python
def remove_khmer_punctuation(text):
    khmer_punct = "។៕!?,៖ៗ"  # Khmer and common punctuation
    cleaned = "".join([c for c in text if c not in khmer_punct])
    cleaned = " ".join(cleaned.split())  # remove extra spaces
    return cleaned

all_sentences_cleaned = []
for s in all_sentences:
    s = keep_only_khmer(s)         # keep Khmer letters + punctuation
    s = remove_khmer_numbers(s)    # remove Khmer numbers
    s = remove_khmer_punctuation(s) # remove punctuation
    s = s.strip()                   # remove extra spaces
    if s:                           # skip empty sentences
        all_sentences_cleaned.append(s)
print("First 100 sentences without numbers and punctuation:")
for i, sent in enumerate(all_sentences_cleaned[:100], 1):
    print(f"{i}: {sent}")
```

5. Removed the short sentences( not meaningful)

```python
def remove_short_sentences(text, min_words=2):
    words = text.split()
    if len(words) < min_words:
        return ""
    return text

all_sentences_cleaned_1 = []

for s in all_sentences_cleaned:
    s = remove_short_sentences(s, min_words=3)  # remove sentences < 3 characters
    if s:
        all_sentences_cleaned_1.append(s)
print("First 100 final cleaned sentences:")
for i, sent in enumerate(all_sentences_cleaned_1[:100], 1):
    print(f"{i}: {sent}")
```

6. Segmentation the from setences to words

```python
from khmernltk import word_tokenize

segmented_sentences = []

for sent in all_sentences_cleaned_1:
    tokens = word_tokenize(sent)  # tokenize into Khmer words
    if tokens:
        segmented_sentences.append(" ".join(tokens))

# Preview
print("First 10 segmented sentences:")
for i, sent in enumerate(segmented_sentences[:100], 1):
    print(f"{i}: {sent}")
```

```python
import pandas as pd

# Define the mapping of original groups to broad classes
BROAD_CLASS_MAPPING = {
    'Conjunctions': 'Conjunctions',
    'Pronouns': 'Pronouns',
    'Determiners & Quantifiers': 'Determiners & Quantifiers',
    'Prepositions / Relational Words': 'Prepositions / Relational Words',
    'Auxiliary Verbs / Aspect Markers': 'Auxiliary Verbs / Aspect Markers',
    'Particles & Discourse Markers': 'Particles & Discourse Markers',
    'Question & Negation Words': 'Question & Negation Words',
    'Function Nouns': 'Function Nouns',
    'Numbers & Time Expressions': 'Numbers & Time Expressions',
    'Politeness & Honorifics': 'Politeness & Honorifics'
}

# Read the CSV file
df = pd.read_csv(r'D:\Year 5\S1\Information Retrieval\StopwordProject\khmer_stopword_project\stopwords\FIle_Stopwords.csv')  # Replace with your actual file path

# Display the first few rows to understand the structure
print("Original data shape:", df.shape)
print("\nFirst few rows:")
print(df.head())

# Clean the data - remove any leading/trailing whitespace
df['term'] = df['term'].str.strip()
df['linguistic_group'] = df['linguistic_group'].str.strip()

# Step 1: Map to broad classes
def map_to_broad_class(original_group):
    # Check if the original group matches exactly with a broad class
    if original_group in BROAD_CLASS_MAPPING:
        return BROAD_CLASS_MAPPING[original_group]
    
    # Check for combined groups (like "Auxiliary Verbs / Aspect Markers + Content word")
    for broad_class in BROAD_CLASS_MAPPING:
        if broad_class in original_group:
            return broad_class
    
    return None

df['broad_group'] = df['linguistic_group'].apply(map_to_broad_class)

# Step 2: Remove terms that don't fall into the 10 classes
df_filtered = df[df['broad_group'].notna()].copy()

# Step 3: Remove duplicate terms (keep first occurrence)
df_unique = df_filtered.drop_duplicates(subset=['term'], keep='first')

# Step 4: Create final dataframe with only term and broad_group
final_df = df_unique[['term', 'broad_group']].copy()

# Sort by term for better organization
final_df = final_df.sort_values('term').reset_index(drop=True)

# Display summary
print(f"\nOriginal records: {len(df)}")
print(f"Records after filtering to 10 classes: {len(df_filtered)}")
print(f"Records after removing duplicates: {len(df_unique)}")

print(f"\nDistribution by broad class:")
print(final_df['broad_group'].value_counts())

print("\nFirst few rows of final dataframe:")
print(final_df.head(20))

# Save to CSV
final_df.to_csv('khmer_stopwords_broad_classes.csv', index=False, encoding='utf-8-sig')
print(f"\nFinal CSV saved with {len(final_df)} unique terms.")
```

```python
# %% [markdown]
# # Remove Stop word Process 

# %%
import pandas as pd
import os
from khmernltk import word_tokenize

# 1. LOAD AND PREPARE STOPWORDS
stopwords_path = r'D:\Year 5\S1\Information Retrieval\StopwordProject\khmer_stopword_project\stopwords\khmer_stopwords_broad_classes.csv'
stopwords_df = pd.read_csv(stopwords_path, encoding='utf-8-sig')

# Create dictionaries for each linguistic group
stopwords_by_group = {}
all_stopwords = set()

for group in stopwords_df['broad_group'].unique():
    group_words = set(stopwords_df[stopwords_df['broad_group'] == group]['term'].str.strip().tolist())
    stopwords_by_group[group] = group_words
    all_stopwords.update(group_words)

print(f"Total stopwords: {len(all_stopwords)}")
print(f"Linguistic groups: {list(stopwords_by_group.keys())}")
print(f"Stopwords per group:")
for group, words in stopwords_by_group.items():
    print(f"  {group}: {len(words)} words")

# 2. USE THE SEGMENTED SENTENCES YOU ALREADY CREATED
# You already have segmented_sentences from your previous code
# Let's verify it exists
try:
    print(f"\nSegmented sentences available: {len(segmented_sentences)}")
    if segmented_sentences:
        print("First 3 segmented sentences:")
        for i, sent in enumerate(segmented_sentences[:3], 1):
            print(f"{i}: {sent}")
except NameError:
    print("ERROR: segmented_sentences is not defined. Please run the segmentation cell first.")
    # Re-run segmentation if needed
    segmented_sentences = []
    for sent in all_sentences_cleaned_1:
        tokens = word_tokenize(sent)
        if tokens:
            segmented_sentences.append(" ".join(tokens))
    print(f"Re-created {len(segmented_sentences)} segmented sentences")

# 3. CREATE OUTPUT DIRECTORY
output_dir = r'D:\Year 5\S1\Information Retrieval\StopwordProject\khmer_stopword_project\data\processed'
os.makedirs(output_dir, exist_ok=True)
print(f"\nOutput directory: {output_dir}")

# 4. DEFINE STOPWORD REMOVAL FUNCTIONS
def remove_stopwords_from_sentence(sentence, stopword_set):
    """Remove stopwords from a single sentence"""
    words = sentence.split()
    filtered_words = [word for word in words if word not in stopword_set]
    return " ".join(filtered_words)

def remove_stopwords_from_all_sentences(sentences, stopword_set):
    """Remove stopwords from all sentences"""
    return [remove_stopwords_from_sentence(sent, stopword_set) for sent in sentences]

# 5. REMOVE STOPWORDS GROUP BY GROUP
results_by_group = {}
files_created = []

print("\n" + "="*60)
print("REMOVING STOPWORDS BY LINGUISTIC GROUP")
print("="*60)

# Process each linguistic group separately
for group_name, group_stopwords in stopwords_by_group.items():
    print(f"\n[{group_name}]")
    print(f"  Stopwords in this group: {len(group_stopwords)}")
    
    # Remove stopwords from this specific group
    filtered_sentences = remove_stopwords_from_all_sentences(segmented_sentences, group_stopwords)
    
    # Store results
    results_by_group[group_name] = filtered_sentences
    
    # Save to file
    safe_name = group_name.replace(' ', '_').replace('/', '_').replace('&', 'and')
    filename = f"no_{safe_name}.txt"
    filepath = os.path.join(output_dir, filename)
    
    with open(filepath, 'w', encoding='utf-8') as f:
        for sentence in filtered_sentences:
            f.write(sentence + '\n')
    
    files_created.append(filename)
    
    # Calculate statistics
    original_words = sum(len(s.split()) for s in segmented_sentences)
    filtered_words = sum(len(s.split()) for s in filtered_sentences)
    removed_count = original_words - filtered_words
    
    print(f"  Saved: {filename}")
    print(f"  Words removed: {removed_count} ({removed_count/original_words*100:.1f}%)")
    
    # Show sample comparison
    if len(segmented_sentences) > 0:
        original_sample = segmented_sentences[0]
        filtered_sample = filtered_sentences[0]
        if original_sample != filtered_sample:
            print(f"  Sample comparison (sentence 1):")
            print(f"    Original: {original_sample[:80]}...")
            print(f"    Filtered: {filtered_sample[:80]}...")
        else:
            print(f"  Note: No stopwords from this group found in first sentence")

# 6. REMOVE ALL STOPWORDS AT ONCE
print("\n" + "="*60)
print("REMOVING ALL STOPWORDS")
print("="*60)

all_filtered_sentences = remove_stopwords_from_all_sentences(segmented_sentences, all_stopwords)

# Save all-filtered results
all_filepath = os.path.join(output_dir, "no_all_stopwords.txt")
with open(all_filepath, 'w', encoding='utf-8') as f:
    for sentence in all_filtered_sentences:
        f.write(sentence + '\n')

files_created.append("no_all_stopwords.txt")

# Calculate statistics
original_words = sum(len(s.split()) for s in segmented_sentences)
all_words = sum(len(s.split()) for s in all_filtered_sentences)
all_removed = original_words - all_words

print(f"  Saved: no_all_stopwords.txt")
print(f"  Total words removed: {all_removed} ({all_removed/original_words*100:.1f}%)")
print(f"  Sample (sentence 1): {all_filtered_sentences[0][:100]}...")

# 7. SAVE ORIGINAL SEGMENTED SENTENCES
original_filepath = os.path.join(output_dir, "original_segmented_sentences.txt")
with open(original_filepath, 'w', encoding='utf-8') as f:
    for sentence in segmented_sentences:
        f.write(sentence + '\n')

files_created.append("original_segmented_sentences.txt")
print(f"\nSaved original segmented sentences")

# 8. CREATE COMPREHENSIVE STATISTICS FILE
print("\n" + "="*60)
print("CREATING STATISTICS SUMMARY")
print("="*60)

# Calculate statistics for each group
stats_data = []
for group_name, filtered_sentences in results_by_group.items():
    word_count = sum(len(s.split()) for s in filtered_sentences)
    removed = original_words - word_count
    percentage = removed/original_words*100
    stats_data.append({
        'Group': group_name,
        'Type': 'Single Group',
        'Words_Remaining': word_count,
        'Words_Removed': removed,
        'Removal_Percentage': percentage
    })

# Add all groups statistics
stats_data.append({
    'Group': 'ALL GROUPS',
    'Type': 'All Groups',
    'Words_Remaining': all_words,
    'Words_Removed': all_removed,
    'Removal_Percentage': all_removed/original_words*100
})

# Create and save statistics
stats_df = pd.DataFrame(stats_data)
stats_filepath = os.path.join(output_dir, "stopword_removal_statistics.csv")
stats_df.to_csv(stats_filepath, index=False, encoding='utf-8-sig')
files_created.append("stopword_removal_statistics.csv")

print(f"\nStatistics saved to: stopword_removal_statistics.csv")
print("\nSummary of removal percentages:")
for _, row in stats_df.iterrows():
    print(f"  {row['Group']}: {row['Removal_Percentage']:.2f}%")

# 9. CREATE COMPARISON FILE (First 10 sentences)
comparison_filepath = os.path.join(output_dir, "comparison_samples.txt")
with open(comparison_filepath, 'w', encoding='utf-8') as f:
    f.write("STOPWORD REMOVAL COMPARISON - FIRST 10 SENTENCES\n")
    f.write("="*80 + "\n\n")
    
    for i in range(min(10, len(segmented_sentences))):
        f.write(f"SENTENCE {i+1}\n")
        f.write(f"Original: {segmented_sentences[i]}\n")
        
        # Show each group's version
        for group_name in stopwords_by_group.keys():
            filtered = results_by_group[group_name][i]
            if filtered != segmented_sentences[i]:
                f.write(f"No {group_name}: {filtered}\n")
        
        # Show all-stopwords version
        f.write(f"No ALL: {all_filtered_sentences[i]}\n")
        f.write("-"*80 + "\n\n")

files_created.append("comparison_samples.txt")

# 10. FINAL SUMMARY
print("\n" + "="*60)
print("PROCESSING COMPLETE - FINAL SUMMARY")
print("="*60)

print(f"\nCorpus Statistics:")
print(f"  Total sentences: {len(segmented_sentences)}")
print(f"  Total words (original): {original_words}")
print(f"  Total words (after all stopwords removed): {all_words}")
print(f"  Total stopwords removed: {all_removed}")
print(f"  Overall reduction: {all_removed/original_words*100:.1f}%")

print(f"\nFiles created in {output_dir}:")
for i, filename in enumerate(files_created, 1):
    filepath = os.path.join(output_dir, filename)
    size_kb = os.path.getsize(filepath) / 1024
    print(f"  {i:2d}. {filename:<35} ({size_kb:.1f} KB)")

print(f"\nFiles by type:")
print(f"  - 10 group-specific filtered files (no_*.txt)")
print(f"  - 1 all-stopwords filtered file (no_all_stopwords.txt)")
print(f"  - 1 original segmented file (original_segmented_sentences.txt)")
print(f"  - 1 statistics CSV file (stopword_removal_statistics.csv)")
print(f"  - 1 comparison samples file (comparison_samples.txt)")

# 11. QUICK ANALYSIS OF MOST COMMON STOPWORDS REMOVED
print("\n" + "="*60)
print("ANALYSIS: Most Frequently Removed Stopwords")
print("="*60)

# Create a frequency counter for stopwords in the corpus
from collections import Counter

stopword_freq = Counter()
all_corpus_words = []

# Collect all words from segmented sentences
for sentence in segmented_sentences:
    words = sentence.split()
    all_corpus_words.extend(words)

# Count stopword occurrences
for word in all_corpus_words:
    if word in all_stopwords:
        stopword_freq[word] += 1

# Get top 20 most frequent stopwords
top_20 = stopword_freq.most_common(20)

print("\nTop 20 most frequent stopwords in your corpus:")
print("-"*50)
for word, count in top_20:
    # Find which group this stopword belongs to
    group = None
    for grp, words_set in stopwords_by_group.items():
        if word in words_set:
            group = grp
            break
    
    print(f"  {word:<15} : {count:>5} occurrences ({group})")

# Save this analysis
analysis_filepath = os.path.join(output_dir, "stopword_frequency_analysis.txt")
with open(analysis_filepath, 'w', encoding='utf-8') as f:
    f.write("STOPWORD FREQUENCY ANALYSIS IN CORPUS\n")
    f.write("="*60 + "\n\n")
    f.write(f"Total unique stopwords: {len(all_stopwords)}\n")
    f.write(f"Total stopword occurrences: {sum(stopword_freq.values())}\n")
    f.write(f"Percentage of corpus that are stopwords: {sum(stopword_freq.values())/len(all_corpus_words)*100:.1f}%\n\n")
    
    f.write("Top 50 most frequent stopwords:\n")
    f.write("-"*60 + "\n")
    for word, count in stopword_freq.most_common(50):
        # Find group
        group = None
        for grp, words_set in stopwords_by_group.items():
            if word in words_set:
                group = grp
                break
        
        f.write(f"{word:<20} {count:>6} ({group})\n")

files_created.append("stopword_frequency_analysis.txt")
print(f"\nFrequency analysis saved to: stopword_frequency_analysis.txt")
```
